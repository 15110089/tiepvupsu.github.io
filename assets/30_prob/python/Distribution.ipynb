```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
from matplotlib.backends.backend_pdf import PdfPages

with PdfPages('uni_norm.pdf') as pdf:
    mu = 0
    variance = 1
    sigma = math.sqrt(variance)
    x = np.linspace(mu-5*variance,mu+5*variance, 200)
    plt.plot(x,mlab.normpdf(x, mu, sigma), 'b')

    mu = 0.5
    variance = 0.5
    sigma = math.sqrt(variance)
    x = np.linspace(mu-6*variance,mu+6*variance, 200)
    plt.plot(x,mlab.normpdf(x, mu, sigma), 'r')

    mu = -1
    variance = 2
    sigma = math.sqrt(variance)
    x = np.linspace(mu-4*variance,mu+4*variance, 200)
    plt.plot(x,mlab.normpdf(x, mu, sigma), 'g')
    plt.xlabel('$x$', fontsize = 15)
    plt.ylabel('$Pr(x)$', fontsize = 15)
    plt.plot([-8, 8], [0, 0], 'k',)

    plt.text(-5.75, .3, r'$\mu = 0, \sigma^2 = 1$', fontsize=15, color = 'blue')
    plt.text(1.25, .5, r'$\mu = .5, \sigma^2 = .5$', fontsize=15, color = 'red')
    plt.text(-8.25, .15, r'$\mu = -1, \sigma^2 = 2$', fontsize=15, color = 'green')
    # plt.text(2, 6)
    pdf.savefig()
    plt.show()
```

```python

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np 



fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(1)

X = np.arange(-2, 2, 0.025)
Y = np.arange(-2, 2, 0.025)
X, Y = np.meshgrid(X, Y)

Z = np.exp(-(X**2 + Y**2 - .5*X*Y))*1.5


ax.plot_surface(X, Y, Z, alpha=1, cmap=cm.jet)
cset = ax.contour(X, Y, Z, zdir='z', offset=-1, cmap=cm.jet)

ax.set_xlabel('$x$', fontsize = 15)
ax.set_xlim(-2, 2)
ax.set_ylabel('$y$', fontsize =15)
ax.set_ylim(-2, 2)

# ax.set_zlabel('$Pr(x, y)$', fontsize = 10)

ax.set_title('bivariate normal distribution', fontsize = 15)
ax.set_zlim(-1, 1)
ax.set_zticks([])
ax.set_xticks([-2,  0,  2])
ax.set_yticks([-2,  0,  2])
# plt.axis('equal')
# ax.auto_scale_xyz([0, 500], [0, 500], [0, 0.15])
plt.savefig('aa.png')
plt.show()
```
